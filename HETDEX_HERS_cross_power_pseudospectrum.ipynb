```python
path_sav = os.path.join(os.environ['MAPSPATH'],'Herschel','bandpowers','combined_spectra_all_corrected_errors')
file_sav = 'combined_spectra_extended_sources_only_dl_160_bin_width_25_percent_'
wv_key = {'17p6': 'PSW', '23p9': 'PMW', '35p2': 'PLW'}
def plot_pseudospectra(self, pk_dict):

    keys_one = np.unique([i.split('x')[0] for i in pk_dict])
    keys_zed = np.unique([i.split('x')[1].split('__')[0] for i in pk_dict])
    keys_mag = np.unique([i.split('x')[1].split('__')[1] for i in pk_dict])
    keys_fwhm = np.unique([i.split('x')[1].split('__')[2] for i in pk_dict])
    
    nx=len(keys_mag)
    ny=len(keys_zed)
    color_wv = ['b','g','r']
    fig, axs = plt.subplots(nx, ny, figsize=(22, 10))
    for ii, key_mag in enumerate(keys_mag):
        for iii, key_zed in enumerate(keys_zed):
            for iwv, key_fwhm in enumerate(keys_fwhm):
                key_wv = wv_key[key_fwhm.split('_')[1]]

                key_one = '_'.join(['hers_act', key_wv.lower()])
                key_two = '__'.join([key_zed, key_mag, key_fwhm])
                key = 'x'.join([key_one,key_two]) 
                #pdb.set_trace()
                
                key_title_one = key.replace('_','-').split('x')[0].split('-')[-1].upper() #+ 'xHETDEX ('
                key_title_two = key.replace('p','.').split('x')[1].replace('_','-').split('--')[0].replace('redshift-','z=')+','
                key_title_two += key.replace('p','.').split('x')[1].replace('_','-').split('--')[1].replace('magnitude-','mag=')#+')'

                if 1: #key in pk_dict:
                    xcorr_label = '{0:0.0f}x{1:0.0f}'.format(pk_dict[key]['wv1'], pk_dict[key]['wv2'])

                    path_v12 = os.path.join(path_sav, file_sav + xcorr_label + '.sav')

                    ell = pk_dict[key]['ell']
                    k_theta = self.ell_to_k(ell)

                    if os.path.isfile(path_v12):
                        x = readsav(path_v12)
                        y = x['combined_spectra'][:, 0, 0]
                        axs[ii, iii].plot(x['k_t'], y, 's', c=color_wv[iwv], label=key_title_one)
                    else:
                        print('No file found {}'.format(path_v12))

                    pk = pk_dict[key]['pk_raw']
                    pk_corrected_psf = pk_dict[key]['pk_beam_corrected']
                    #pdb.set_trace()


                    # HETDEX x HerS
                    plot_mkk = False
                    if 'pk_mkk_corrected' in pk_dict[key]:
                        pk_corrected_mkk = pk_dict[key]['pk_mkk_corrected']
                        pk_corrected_both = pk_dict[key]['pk_mkk_and_beam_corrected']
                        plot_mkk = True

                    #axs[ii, iii].plot(k_theta, pk, '-o', c='g', label='raw')
                    #axs[ii, iii].plot(k_theta, pk_corrected_psf, '-o', c='r', label='beam corr')
                    if plot_mkk:
                        y_Hxh = (np.sqrt(pk_corrected_both)**2)
                        y_Hxh_n = (np.sqrt(-pk_corrected_both)**2)
                        axs[ii,iii].plot(k_theta, y_Hxh,'-o', lw=2, c=color_wv[iwv], label='{} X hetdex'.format(key_wv));
                        axs[ii,iii].plot(k_theta, y_Hxh_n,'--x', lw=2, c=color_wv[iwv]);

                    axs[ii, iii].set_xlim([0.005, 2.05])
                    axs[ii, iii].set_ylim([1e0, 3e5])
                    axs[ii, iii].set_xscale('log')
                    axs[ii, iii].set_yscale('log')

                    axs[ii, iii].set_title(key_title_two)

                    if not iii:
                        axs[ii, iii].set_ylabel('P(k_theta) (Jy^2 sr^-1)')
                    else:
                        axs[ii, iii].set_yticklabels([])
                    if ii < (nx-1):
                        axs[ii, iii].set_xticklabels([])
                    else:
                        axs[ii, iii].set_xlabel('k_theta (arcmin^-1)')
                    if iii == (ny-1):
                        axs[ii, iii].set_ylabel(key_title_two)
                        axs[ii, iii].yaxis.set_label_position("right")

                    axs[ii, iii].legend(loc='upper right', fontsize=8)
```

```python
path_sav = os.path.join(os.environ['MAPSPATH'],'Herschel','bandpowers','combined_spectra_all_corrected_errors')
file_sav = 'combined_spectra_extended_sources_only_dl_160_bin_width_25_percent_'

def plot_pseudospectra(self, pk_dict):

    keys_one = np.unique([i.split('x')[0] for i in pk_dict])
    keys_two = np.unique([i.split('x')[1] for i in pk_dict])
    
    nx=len(keys_one)
    ny=int(len(keys_two)/len(keys_one))
    fig, axs = plt.subplots(nx, ny, figsize=(22, 10))
    for ii, key_one in enumerate(keys_one):
        iii = 0
        for j, key_two in enumerate(keys_two):
            key = 'x'.join([key_one,key_two]) 
            
            if key in pk_dict:
                xcorr_label = '{0:0.0f}x{1:0.0f}'.format(pk_dict[key]['wv1'], pk_dict[key]['wv2'])

                path_v12 = os.path.join(path_sav, file_sav + xcorr_label + '.sav')
                
                ell = pk_dict[key]['ell']
                k_theta = self.ell_to_k(ell)
                
                if os.path.isfile(path_v12):
                    x = readsav(path_v12)
                    y = x['combined_spectra'][:, 0, 0]
                    axs[ii, iii].plot(x['k_t'], y, 's', c='c', label='V12')
                else:
                    print('No file found {}'.format(path_v12))
                    
                pk = pk_dict[key]['pk_raw']
                pk_corrected_psf = pk_dict[key]['pk_beam_corrected']
                #pdb.set_trace()

                plot_mkk = False
                if 'pk_mkk_corrected' in pk_dict[key]:
                    pk_corrected_mkk = pk_dict[key]['pk_mkk_corrected']
                    pk_corrected_both = pk_dict[key]['pk_mkk_and_beam_corrected']
                    plot_mkk = True

                #axs[ii, iii].plot(k_theta, pk, '-o', c='g', label='raw')
                axs[ii, iii].plot(k_theta, pk_corrected_psf, '-o', c='r', label='beam corr')
                if plot_mkk:
                    #axs[ii, iii].plot(k_theta, pk_corrected_mkk, '-o', c='b', label='mkk corr')
                    axs[ii, iii].plot(k_theta, pk_corrected_both, '-o', c='g', label='beam corr')

                axs[ii, iii].set_xlim([0.005, 2.05])
                axs[ii, iii].set_ylim([1e0, 3e5])
                axs[ii, iii].set_xscale('log')
                axs[ii, iii].set_yscale('log')
                key_title_one = key.replace('_','-').split('x')[0].split('-')[-1].upper() #+ 'xHETDEX ('
                key_title_two = key.replace('p','.').split('x')[1].replace('_','-').split('--')[0].replace('redshift-','z=')+','
                key_title_two += key.replace('p','.').split('x')[1].replace('_','-').split('--')[1].replace('magnitude-','mag=')#+')'
                    
                axs[ii, iii].set_title(key_title_two)

                if not iii:
                    axs[ii, iii].set_ylabel('P(k_theta) (Jy^2 sr^-1)')
                else:
                    axs[ii, iii].set_yticklabels([])
                if ii < (nx-1):
                    axs[ii, iii].set_xticklabels([])
                else:
                    axs[ii, iii].set_xlabel('k_theta (arcmin^-1)')
                if iii == (ny-1):
                    axs[ii, iii].set_ylabel(key_title_one)
                    axs[ii, iii].yaxis.set_label_position("right")
                
                axs[ii, iii].legend(loc='upper right', fontsize=8)
                iii+=1
```
